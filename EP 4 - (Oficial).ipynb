# Análise de dados dos eleitores de SP

Com intuito de aprendizagem com a utilização da linguagem Python para analisar dados públicos, realizamos uma pesquisa baseasa nos dados dos eleitores do estado de São Paulo, nas eleições de 2020.

## 1 - Importação das bibliotecas

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

## 2 - Criação do dataframe com as colunas utilizadas do arquivo

```python
df = pd.read_csv(r'C:\Users\Feh\Desktop\EP4\perfil_eleitor_secao_ATUAL_SP.csv', delimiter = ';', encoding = 'iso-8859-1', usecols = ['SG_UF', 'CD_GENERO', 'DS_GENERO', 'CD_ESTADO_CIVIL', 'DS_ESTADO_CIVIL',
                                                                                                                                     'CD_FAIXA_ETARIA', 'DS_FAIXA_ETARIA', 'CD_GRAU_ESCOLARIDADE', 'DS_GRAU_ESCOLARIDADE'])
```

```python
df.sample(10)
```

## 3 - Criação do gráfico comparativo de gênero dos eleitores

## 3.1 - Gráfico em barras com valor total

```python
colors = ['#eb498c', '#5EF38C']
df.groupby('DS_GENERO').size().sort_values().plot(kind='barh',stacked=True , title='Gênero dos eleitores de SP', color=colors, figsize=(8, 5))
```

## 3.2 - Gráfico pizza com porcentagem

```python
df.groupby('DS_GENERO').size().sort_values().plot(kind='pie', autopct='%.2f%%', figsize=(8, 5), title='Gênero dos eleitores de SP')
```

## 4 - Criação do gráfico comparativo de grau de escolaridade dos eleitores

```python
df.groupby('DS_GRAU_ESCOLARIDADE').size().sort_values().plot(kind='barh', title='Grau de Escolaridade eleitores de SP', figsize=(8, 5))
```

## 5 - Criação do gráfico comparativo da faixa etária dos eleitores

```python
df.groupby('DS_FAIXA_ETARIA').size().sort_values().plot(kind='barh', title='Faixa Etária dos eleitores de SP', figsize=(8, 8))
```

## 6 - Criação do gráfico comparativo de estado civil dos eleitores

```python
df.groupby('DS_ESTADO_CIVIL').size().sort_values().plot(kind='barh', title='Estado Civil dos eleitores de SP', figsize=(8, 5))
```

## 7 - Criação do gráfico comparativo de gênero por faixa etária

```python
df['GENERO_FAIXA_ET'] = df['DS_GENERO'] + ' ' + df['DS_FAIXA_ETARIA']
df.groupby('GENERO_FAIXA_ET').size().sort_values().plot(kind='barh', title='Relação Gênero/Faixa Etária dos eleitores de SP',color=colors, figsize=(12, 15))
```

## 8 - Criação do gráfico comparativo de Grau de Escolaridade por Gênero

```python
df['GENERO_GRAU_ESC'] = df['DS_GENERO'] + ' ' + df['DS_GRAU_ESCOLARIDADE']
df.groupby('GENERO_GRAU_ESC').size().sort_values().plot(kind='barh', title='Relação Gênero/Faixa Etária dos eleitores de SP', figsize=(12, 15))
```

## Conclusão

O número de eleitores do gênero feminino é superior ao número de eleitores do gênero masculino, por consequência os números totais na comparação de grau de escolaridade também são maiores referentes ao gênero feminino, porém pelo gráfico comparativo de grau de escolaridade por gênero, podemos concluir que mulheres, no geral, tem um grau de escolaridade maior, mostrando a busca pelo crescimento profissional.
